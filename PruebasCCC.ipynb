```python
size = int(input())
map = [input() for i in range(size)]
nCoordenadas = int(input())

coordenadas = [map(int, input().split(",")) for i in range(nCoordenadas)]

salida = ""

for i in coordenadas:
    salida+=map[i[0]][i[1]]+"\n"

print(salida)
```

##**Vectores **##

```python
vector_size = int(input())
v, w = None, None

try:
    v, w = list(map(float, input().split(" "))), list(map(float, input().split(" ")))
    if not len(v)==len(w):
        raise Exception()
except:
    print("BOBO")

s = sum(i[0]*i[1] for i in zip(v, w))
print("\n".join(map(str, [v, w, s])))
```

```python
import numpy as np

vector_size = int(input())
v, w = None, None
try:
    v, w = np.array(input().split(" "), dtype=np.float128
                    ), np.array(input().split(" "), dtype=np.float128)
    if not len(v)==len(w):
        raise Exception()
except:
    print("Error en la entrada")

s = np.dot(v, w)
print("\n".join(map(str, [v, w, s])))
```
